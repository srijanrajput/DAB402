```python
output_df= pd.DataFrame(columns= ['Count','NA','Null','Unique','dtype','Mean','Mode','Std','Min','Quantile-25%','Median','quantile-75%','Max','Skewness'])
def Univariant_statistics(df):
    for col in df:
        if pd.api.types.is_numeric_dtype(df[col]):
             output_df.loc[col] = [df[col].count(),df[col].isna().sum(),df[col].isnull().sum(),df[col].nunique(),
                                  df[col].dtype,df[col].mean(),df[col].mode().values[0],df[col].std(),df[col].min(),
                                  df[col].quantile(0.25),df[col].median(),df[col].quantile(0.75),df[col].max(),df[col].skew(axis=0,skipna=True)]
        else:
            output_df.loc[col] = [df[col].count(),df[col].isna().sum(),df[col].isnull().sum(),df[col].nunique(),
                                  df[col].dtype,'-','-','-','-','-','-','-','-','-']
    return output_df

Univariant_statistics(df)
```

## KNN

### train test split

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
dfX = dfnew[dfnew.columns[~dfnew.columns.isin(['Education','marital_Status', 'dt_Customer'])]]
dfY = df["Response"]
dfY = dfY.astype(int) # converting response variable to int other KNN will treat this as continuous 
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfX,dfY, test_size=0.2)
```

### Choosing k-value

```python
error_rate=[]
for i in range(1,50): # sqrt of rows is 47.3
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test)
    error_rate.append(np.mean(pred!=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('Error Rate vs K value')
plt.xlabel('K')
plt.ylabel('Error rate')
```

### checking with k=10

```python
knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
pred_k=knn.predict(X_test)
pred_k
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(confusion_matrix(y_test,pred_k))
print(classification_report(y_test,pred_k))
print(accuracy_score(y_test,pred_k)*100)
```

### Decision tree

```python
df
```

```python
from sklearn import tree
from sklearn import metrics
from sklearn.tree import plot_tree


dfX=df[df.columns[~df.columns.isin(['Education','marital_Status','dt_Customer','Response'])]]

X_train, X_test, y_train, y_test = train_test_split(dfX,dfY, test_size=0.1)
# gini_impurity,defines the impurity in the dataset.
model=tree.DecisionTreeClassifier(max_depth=2)

d_tree_fit=model.fit(X_train,y_train)


```

```python
d_tree_predict=d_tree_fit.predict(X_test)
confusion_matrix(y_test,d_tree_predict)
print(classification_report(y_test,d_tree_predict))
print( metrics.accuracy_score(y_test,d_tree_predict)*100)
```

### Plot the decision tree

```python
dec_tree = plot_tree(decision_tree=d_tree_fit, feature_names = dfnew.columns, 
                      filled = True , precision = 4, rounded = True)
```

### using entropy

```python
## criterion for selection measure.Entropy defines the degree of randomness.We always prefer to chose high information gain with low entropy.
d_tree_entropy=tree.DecisionTreeClassifier(criterion="entropy",max_depth=2)

## train decision tree classifier
d_tree_fit_entropy=d_tree_entropy.fit(X_train,y_train)

## predict with testing set
d_tree_predict=d_tree_fit_entropy.predict(X_test)

##check with the confusion ,matrix
confusion_matrix(y_test,d_tree_predict)
print(classification_report(y_test,d_tree_predict))

## check how often your selected classifier is correct
print("Accuracy : ", metrics.accuracy_score(y_test,d_tree_predict)*100)
```

```python
dec_tree = plot_tree(decision_tree=d_tree_fit_entropy, feature_names = dfnew.columns, 
                      filled = True , precision = 6, rounded = True)
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
#dfX = dfnew[dfnew.columns[~dfnew.columns.isin(['Education','Marital_Status', 'Dt_Customer'])]]
data=df[df.columns[~df.columns.isin(['Education','marital_Status','dt_Customer'])]]
```

```python
kmeans=KMeans(n_clusters=3,init='random',n_init=10,max_iter=200,random_state=34)
kmeans
kmeans.fit(data)
cluster_means=kmeans.predict(data)
print("SSE Value:",kmeans.inertia_)    #  defines the lowest SSE Value. SSE is a sum of squared error which always need to minimum .
print("location of centroid :", kmeans.cluster_centers_)   # sepecifies the location of centroid
```

### Choosing appropriate number of clusters

### Elbow Method

```python
kmeans_func={
    "init":"random",
    "n_init":10,
    "max_iter":200,
    "random_state":42
}

sse=[]

for k in range(1,10):
    kmeans_optimal=KMeans(n_clusters=k,**kmeans_func) ## here ** is a dictionary unpacking operator.
    kmeans_optimal.fit(data)
    sse.append(kmeans_optimal.inertia_)
```

```python
kmeans_optimal.inertia_
```

```python
plt.plot(range(1,10),sse)
plt.xlabel(" Number of Clusters")
plt.ylabel("Sum of Squared Error")
plt.show()
```
